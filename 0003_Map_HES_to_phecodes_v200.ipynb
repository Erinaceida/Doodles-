# Phecodes

## Map to phecode v200

```python
# Map pivoted HES APC data
# to phecodes and write out
# long file of events.

phedata = spark.sql("""
 SELECT 
  h.person_id_deid, 
  h.date,
  h.n,
  h.DIAG_4 AS source_code,
  h.DIAG_4 AS DIAG_4,
  p.phecode,
  'hes_apc'  AS source
FROM
  dars_nic_391419_j3w9t_collab.ccu013_apc_pivot h
LEFT JOIN
  dsa_391419_j3w9t_collab.ccu013_lkp_phev200_map p
ON 
  h.diag_4 = p.icd10
WHERE phecode IS NOT NULL
AND YEAR(date) >= 1998
""")

phedata.createOrReplaceTempView('ccu013_phedatav200')
display(phedata)
```
