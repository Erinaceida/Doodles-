## Create temp table with lkp phe_dict and lkp phe_speciality

```python
dict_spe = spark.sql("""
select d.phecode, d.type, d.level, d.f_disease_node, d.sex, s.spec_01 as speciality
from dsa_391419_j3w9t_collab.ccu013_lkp_phev200_phedict d  
left join dsa_391419_j3w9t_collab.ccu013_lkp_phev200_speciality s
on d.phecode = s.phecode """)
```

```python
dict_spe.createOrReplaceTempView('ccu013_lkp_phev200_temp_phedict_spe')
```

```python
drop_table("ccu013_lkp_phev200_temp_phedict_spe")
```

```python
dict_spe.write.mode('overwrite').saveAsTable('dsa_391419_j3w9t_collab.ccu013_lkp_phev200_temp_phedict_spe')
```
